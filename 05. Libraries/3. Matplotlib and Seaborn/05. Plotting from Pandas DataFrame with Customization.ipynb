### Which one should you use?(pyplot vs matplotlib oo method?)

* When Plotting Something Quickly, use pyplot method
* When Plotting Something more Advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot Method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO Method (Object Oriented) mixed with pyplot

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax = ax);
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax = ax);
ax.set_xlim([45,100]);
```

```python
# OO Method From Scratch
fig, ax = plt.subplots(figsize=(10,6))

#Plot the Data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Customise the plot
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel = "Cholestrol")

# Add a Legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a Horizontal Line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of Chol, Age, Thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Level",
       xlabel="Age",
       ylabel="Cholestrol")
# Add a Legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--");

# Add Data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# Customize ax1

ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

# Add a Legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a Title to the Figure
fig.suptitle("Heart Disease Analysis", fontsize=18, fontweight="bold");
```

### Customizing the X and Y axis Limitations

```python
# Subplot of Chol, Age, Thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Level",
       ylabel="Cholestrol")

# Change the X-Axis Limits
ax0.set_xlim([50,80])

# Add a Legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--");

# Add Data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")
# Customize ax1

ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
# Change the X-Axis Limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

# Add a Legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a Title to the Figure
fig.suptitle("Heart Disease Analysis", fontsize=18, fontweight="bold");
```
