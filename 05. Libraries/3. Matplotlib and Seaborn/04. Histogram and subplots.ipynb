# Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Make a plot from Dictionary

nut_butter_prices = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = " Nut Butter Store", ylabel = "Price ($)");
```

```python
# Horizontal Bar Graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some Data and plot Histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two Options for Subplots

```python
# Subplots Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each Different Axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.hist(np.random.randn(1000))
ax4.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Subplots option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each Different Axes Index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].hist(np.random.randn(1000))
ax[1, 1].barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```
